# Regression linéaire et méthode des moindres carrés

## Introduction
La régression linéaire est l'action de modéliser la relation entre deux variables 
$x$ (variable indépendante) et $y$ (variable dépendante) par une droite d'équation $y=ax+b$. La méthode des moindres carrés est une méthode parmi d'autres pour effectuer une régression linéaire. Si on suppose qu'on a obtenu lors d'une mesure $N$ couples de points expérimentaux $(x_i ; y_i)$, la méthode des moindres carrés minimise les écarts entre les valeurs mesurées $y_i$ et les valeurs calculées $ax_i+b$.

L'image ci-dessous vous montre une série de données (points noirs), la droite théorique (qu'on cherche à obtenir) qui passe au plus près de ces points (en rouge). L'objectif est de calculer les coefficients $a$ (la pente ou coefficient directeur) et $b$ (l'ordonnée à l'origine) de cette droite.

Sur la figure, on a représenté en bleu les écarts entre l'ordonnée d'un point expérimental $y_i$ et l'ordonnée correspondante calculée $y=ax_i+b$. Rappelez-vous du problème rencontré si on calcule la moyenne des écarts à la moyenne: elle est nulle. Nous avons donc dû introduire la variance et l'écart-type pour obtenir des valeurs significatives d'une incertitude sur un (grand ?) nombre de données. 

En pratique, pour trouver les coefficients $a$ et $b$ de la droite qui passe au plus près des $N$ points de mesure, on se base sur le même principe : on minimise la somme des **carrés** des écarts représentés en bleu. On doit donc minimiser la fonction :

<p style="text-align:center;">$S(a,b)=\displaystyle\sum_{i=1}^N[y_i-(ax_i+b)]^2$</p>

La minimisation se fait en trouvant les valeurs de $a$ et $b$ qui annulent les dérivées partielles de $S(a,b)$ :

<p style="text-align:center;">$\dfrac{\partial S(a,b)}{\partial a}=0$ et $\dfrac{\partial S(a,b)}{\partial b}=0$</p>

Pour un rappel sur les dérivées partielles, voir le document correspondant dans le cours Ametice.

## Calcul des coefficients $a$ et $b$
Soit $S(a,b)$ la somme des carrés des écarts, donnée par

<p style="text-align:center;">$S(a,b)=\displaystyle\sum_{i=1}^N (y_i-ax_i-b)^2$</p>

On dérive cette expression par rapport à $a$ (ce n'est pas habituel, mais dans ce cas, c'est bien $a$ la seule variable considérée, les autres quantités sont considérées comme constantes) : 

$\dfrac{\partial S(a,b)}{\partial a}=\displaystyle\sum_{i=1}^N -2x_i(y_i-ax_i-b)$

Pour minimiser $S$ on doit avoir $\dfrac{\partial S(a,b)}{\partial a}=0$

D'où l'équation (1) :
\begin{equation}
\displaystyle\sum_{i=1}^N x_i(y_i-ax_i-b)=0
\end{equation}

On a une équation, mais deux paramètres à déterminer. On fait donc la même chose avec $b$ : on dérive $S(a,b)$ cette fois par rapport à $b$.

$\dfrac{\partial S(a,b)}{\partial b}=\displaystyle\sum_{i=1}^N -2(y_i-ax_i-b)$

Avec la condition $\dfrac{\partial S(a,b)}{\partial b}=0$, on obtient $\displaystyle\sum_{i=1}^N -2(y_i-ax_i-b)=0$ d'où l'équation (2) :
\begin{equation}
\displaystyle\sum_{i=1}^N (y_i-ax_i-b)=0
\end{equation}

On doit calculer $a$ et $b$ à partir de ce système de deux équations. 

Il est simple de calculer $b$ à partir de l'équation (2) :

$\displaystyle\sum_{i=1}^N (y_i-ax_i-b)=0\Longleftrightarrow\displaystyle\sum_{i=1}^N (y_i-ax_i)-Nb=0$

On peut alléger l'écriture en omettant les bornes de la somme, toujours prise sur les $N$ valeurs mesurées, et donc :

\begin{equation}
b=\dfrac 1N \left(\sum y_i-a\sum x_i\right)
\end{equation}

En remarquant que $\dfrac{\sum_{i=1}^N x_i}{N}=\bar x$ et $\dfrac{\sum_{i=1}^N y_i}{N}=\bar y$ sont les valeurs moyennes respectives de $x_i$ et $y_i$, on trouve donc :

$b=\bar y - a\bar x$

On doit maintenant chercher $a$. Développons l'équation (1) :

(1) $\Longrightarrow\sum x_i\;y_i-\sum  ax_i^2-b\sum x_i=0$

En remplaçant $b$ dans cette équation, on obtient

$\sum x_i\;y_i-a\sum x_i^2-\dfrac 1N \sum x_i \sum y_i+\dfrac 1N a\left(\sum x_i\right)^2=0$

d'où

$N\sum x_i\;y_i-\sum x_i \sum y_i=a\sum x_i^2-a\left(\sum x_i\right)^2$

On en déduit
\begin{equation}
a=\dfrac{N\sum x_i\;y_i-\sum x_i \sum y_i}{\sum x_i^2-\left(\sum x_i\right)^2}\end{equation}


## Exercice résolu : loi de Hooke

|   5   |        10        |          58,6          |
| $N=5$ | $\sum m_i=30$    | $\sum \ell_i=256,6$    |

Elle trace sur un graphique les données "Masse" en abscisse et "Longueur" en ordonnée. Elle espère obtenir une droite dont l'équation lui permettra ensuite de déterminer toute masse inconnue connaissant la longueur correspondante.

Pour calculer l'équation de la droite, elle utilise la méthode des moindres carrés. Elle doit calculer les grandeurs qui lui sont nécessaires : la moyenne des $x$, c'est-à-dire des masses, la moyenne des $y$, c'est-à-dire des longueurs, la moyenne du carré des $x$ et la moyenne du produit des $x$ par les $y$. Elle ajoute donc deux colonnes et une ligne à son tableau pour faire apparaitre les calculs.
    
|       | Masse $m_i$ (kg) | Longueur $\ell_i$ (cm) | $m_i^2$ (kg) | $m_i\,\ell_i$ (kg.cm) |

La moyenne des masses vaut $\overline m = 6$ kg.

La moyenne des longueurs vaut $\bar \ell =\dfrac{256,6}{5} = 51,32$ cm.

Elle calcule alors 

$a=\dfrac{5\times 1622-30 \times 256,6}{5\times 220-30^2}=2,06$ cm/kg

## Tracé des points sur le graphe

```python
#code du TP de Hooke 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def tracer_graphe(fichier):
    """
    Lit les données dans un fichier Excel et trace les points 
    avec les masses comme abscisses et les allongements
    comme ordonnées.

    Args:
    fichier (str): Nom du fichier Excel contenant les données.
    """
    # Lecture du fichier Excel
    try:
        données = pd.read_excel(fichier)
    except FileNotFoundError:
        print(f"Erreur : le fichier {fichier} est introuvable.")
        return
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier : {e}")
        return
    
    # Vérification des colonnes attendues
    if 'masse' not in données.columns or 'longueur' not in données.columns:
        print("Erreur : le fichier doit contenir les colonnes 'masse' et 'longueur'.")
        return
    
    # Extraction des données
    x = données['masse'].tolist()
    y = données['longueur'].tolist()

    #print(x)
    #print(y)
    
    # Tracé des points
    plt.scatter(x, y, color='blue', label='Données')
     
   # Décorations du graphique
    plt.xlabel("Masse (kg)")
    plt.ylabel("Allongement (cm)")
    plt.title('Loi de Hooke')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Axe des x
    plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Axe des y
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
fichier = "mesures.xlsx"
tracer_graphe(fichier)
```

```python
#nouveau code du TP de Hooke 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def tracer_graphe_reg(fichier):
    """
    Lit les données dans un fichier Excel et trace les points 
    avec les masses comme abscisses et les allongements
    comme ordonnées.

    Args:
    fichier (str): Nom du fichier Excel contenant les données.
    """
    # Lecture du fichier Excel
    try:
        données = pd.read_excel(fichier)
    except FileNotFoundError:
        print(f"Erreur : le fichier {fichier} est introuvable.")
        return
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier : {e}")
        return
    
    # Vérification des colonnes attendues
    if 'masse' not in données.columns or 'longueur' not in données.columns:
        print("Erreur : le fichier doit contenir les colonnes 'masse' et 'longueur'.")
        return
    
    # Extraction des données
    x = données['masse'].tolist()
    y = données['longueur'].tolist()

    # Tracé des points
    plt.scatter(x, y, color='blue', label='Données')

    # Calcul des coefficients de régression avec numpy.polyfit
    a, b = np.polyfit(x, y, 1)

    # Générer des points pour tracer la droite
    x_min, x_max = min(x), max(x)
    x_vals = np.linspace(0, x_max, 100)
    y_vals = a * x_vals + b

    # Tracé de la droite de régression
    plt.plot(x_vals, y_vals, color='red', label=f'Droite de régression polyfit \ny = {a:.2f}x + {b:.2f}')

    #Limites du graphe
    plt.xlim(0, x_max+(x_max-x_min)/10) 
    plt.ylim(b-a*(x_max-x_min)/10, max(y_vals)+a*(x_max-x_min)/10) 
    
   # Décorations du graphique
    plt.xlabel("Masse (kg)")
    plt.ylabel("Allongement (cm)")
    plt.title('Loi de Hooke')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Axe des x
    plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Axe des y
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
fichier = "mesures.xlsx"
tracer_graphe_reg(fichier)
    
```
